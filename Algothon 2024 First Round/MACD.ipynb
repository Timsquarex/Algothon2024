```python
import pandas as pd
import numpy as np
```

```python
macd_price_data = pd.read_csv("prices.txt", sep='\s+', header=None, index_col=None)
```

```python
ema9 = macd_price_data.iloc[:,0:50].ewm(span=9, adjust=False).mean()
ema21 = macd_price_data.iloc[:,0:50].ewm(span=21, adjust=False).mean()
```

```python
MACD = ema9 - ema21
MACD
```
